```python
# ...existing code...
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import joblib

def build_random_forest(preprocessor, n_estimators: int = 200, max_depth: int | None = None,
                        random_state: int = 42, n_jobs: int = -1):
    clf = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        random_state=random_state,
        n_jobs=n_jobs
    )
    return Pipeline([("preprocessor", preprocessor), ("clf", clf)])

def build_logistic_regression(preprocessor, C: float = 1.0, max_iter: int = 1000, random_state: int = 42):
    clf = LogisticRegression(C=C, max_iter=max_iter, random_state=random_state)
    return Pipeline([("preprocessor", preprocessor), ("clf", clf)])

def train_model(model, X_train, y_train):
    """Fit the pipeline and return the fitted pipeline."""
    model.fit(X_train, y_train)
    return model

def save_model(model, path):
    joblib.dump(model, path)

def load_model(path):
    return joblib.load(path)
# ...existing code...
```
